```python
os.makedirs('steered_output_eval/judgments', exist_ok=True)

accuracies = {}

for i in range(1, 1):
    file_path = f'steered_output_eval/edit_task_steered_{model.replace("/", "_")}_0.01_iter{i}.json'
    judgment_file_path = f'steered_output_eval/judgments/judge_edit_task_steered_{model.replace("/", "_")}_0.01_iter{i}.json'
    
    judgements = judge_edits(file_path, "gemini")
    
    with open(judgment_file_path, 'w', encoding='utf-8') as f:
        json.dump(judgements, f, indent=4)
    
    counts = extract_verdict_counts(judgements)
    
    # Calculate accuracy (proportion of 1s)
    total = len(judgements)
    accurate = counts.get(1, 0)
    accuracy = accurate / total if total > 0 else 0
    
    # Store in accuracies dict
    accuracies[f'iter{i}'] = {
        'counts': counts,
        'accuracy': accuracy
    }

# Save the accuracies to a separate json file
with open('steered_output_eval/accuracies_edit_lama_3.1_8b.json', 'w', encoding='utf-8') as f:
    json.dump(accuracies, f, indent=4)

print("Judgments and accuracies saved.")
```

```python
import os
import json
from collections import Counter
import re

def extract_verdict_counts_classify(judgments):
    verdicts = []
    for item in judgments:
        judgment_text = item.get('classification', '')
        match = re.search(r'FINAL ANSWER:\s*([-\d]+)', judgment_text)
        if match:
            verdict = int(match.group(1))
            verdicts.append(verdict)
    counts = Counter(verdicts)
    return dict(counts)

os.makedirs('steered_output_eval/classify_results', exist_ok=True)

classify_accuracies = {}

for filename in os.listdir('steered_output_eval'):
    if filename.startswith('classify_task_steered') and filename.endswith('.json'):
        file_path = os.path.join('steered_output_eval', filename)

        with open(file_path, 'r', encoding='utf-8') as f:
            classify_data = json.load(f)
        
        counts = extract_verdict_counts_classify(classify_data)
        
        total = sum(counts.values())
        accurate = counts.get(1, 0)
        accuracy = accurate / total if total > 0 else 0
        
        classify_accuracies[filename] = {
            'counts': counts,
            'accuracy': accuracy,
            'total': total
        }
        
        print(f"{filename}:")
        print(f"  Counts: {counts}")
        print(f"  Accuracy: {accuracy:.4f}")
        print()

with open('steered_output_eval/classify_results/accuracies_classify.json', 'w', encoding='utf-8') as f:
    json.dump(classify_accuracies, f, indent=4)

print("All classify task results saved to steered_output_eval/classify_results/accuracies_classify.json")
```
