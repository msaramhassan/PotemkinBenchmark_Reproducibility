```python
import json
import csv

# Function to convert JSON to CSV
def json_to_csv(json_file, csv_file):
    with open(json_file, 'r') as f:
        data = json.load(f)

    # Open the CSV file for writing
    with open(csv_file, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)

# Example usage
json_file = 'Procedure2_Replication/AutomaticEval/Procedure2_AutomaticEval_Cross-Judge_Results/Cross-Judge Results/deepseek-ai_DeepSeek-R1-Distill-Qwen-7B__judged_by__deepseek-ai_DeepSeek-R1-Distill-Qwen-7B/mmlu/experiment_log_20260322_222455.json'  # Path to your JSON file
csv_file = 'output.csv'   # Path where the CSV should be saved

json_to_csv(json_file, csv_file)
```
